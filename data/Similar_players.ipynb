```
df5 = df4.copy()
df5['right_foot'] = df5['Preffered_Foot'].apply(lambda x: 1 if (x=='Right') else 0)
df5.head()
```

```
print(len(df5['Preffered_Position'].unique()))
positions = df5['Preffered_Position'].unique()
positions
```

We can see there are 200 positions but these are combinations, we can separate out each individual positions and create a multi-hot encoded vector since the position variable is nominal categorial variable.

- Extracting the inidividual positions first

```
all_positions = []
for pos in positions:
    tokens = pos.split('/')
    all_positions.extend(tokens)

unique_positions = list(set(all_positions))
print(len(unique_positions))
unique_positions.remove('GK')
unique_positions = sorted(unique_positions)
print(unique_positions)
```

We can see now we have only 15 positions instead of 200+ positions. Based on this we will hot encode the positions for each player based on their preferences.

Also before one hot encoding, to avoid dummy variable trap we will drop the position GK since it is not seen in combination, and also it is true in real, player who plays as a goal keeper, rarely play in other positions. 

```
position_encoding = {}
for index, pref_pos in zip(df5.index,df5['Preffered_Position'].values):
    multi_hot_encoding = [0]*14
    tokens = pref_pos.split('/')
    for i,pos in enumerate(unique_positions):
        if pos in tokens:
            multi_hot_encoding[i] = 1

    position_encoding[index] = multi_hot_encoding
```

```
pos_encodings = pd.DataFrame.from_dict(data=position_encoding,orient='index',columns=unique_positions)
pos_encodings.head()
```

### 5.3 Cosine Similarities

[Explanation on cosine similarities](https://www.machinelearningplus.com/nlp/cosine-similarity/)

```
def cosine_similarity(vec1,vec2):
    a = np.array(vec1).ravel()
    b = np.array(vec2).ravel()
    denom = np.sqrt(np.sum(np.square(vec1))) * np.sqrt(np.sum(np.square(vec2)))
    return np.dot(a,b) / denom
```

```
cosine_similarity(vectors[0],vectors[1])
```

```
from sklearn.metrics import pairwise as pw

index = 0
def similar_players(index=0,top=10):
    similarity_scores = {}
    for k,v in vectors.items(): 
        similarity_scores[k] = cosine_similarity(vectors[index],v)
    similarity_scores = sorted(similarity_scores.items(), key = lambda item: item[1], reverse=True)
    return similarity_scores[1:top+1]
```

```
basic_info2[basic_info2.iloc[:,0]=='De Gea'].index.values[0]
```

```
player = np.random.randint(0,100)
top_sim_players = similar_players(index=player,top=50)
print('Players similar to:',df9.loc[player,'Name'],":\n")


top_players = {}
for index,similarity in top_sim_players:
    info = list(basic_info2.loc[index,['Name','Age','Preffered_Position','Rating']].values)
    info.insert(len(info),similarity)
    top_players[index] = info

output = pd.DataFrame.from_dict(data=top_players, orient='index',columns=['Name','Age','Preferred Position','Rating','Similarity'])
output.sort_values(by=['Similarity'],ascending=False)
```
